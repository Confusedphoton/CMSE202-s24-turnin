# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

```python
X_22 = sm.add_constant(whr_22_data['healthy_life_expectancy'])
y_22 = whr_22_data['happiness_score']
X_23 = sm.add_constant(whr_23_data['healthy_life_expectancy'])
y_23 = whr_23_data['happiness_score']


model_22 = sm.OLS(y_22, X_22)
results_22 = model_22.fit()

model_23 = sm.OLS(y_23, X_23)
results_23 = model_23.fit()

print(results_22.summary())
print('--'*50)
print(results_23.summary())
```

&#9989;  **Question 3.2 (6 points)**: Answer the following questions: 

1. What were the equations of the best fit lines for 2022 and 2023? 

2. Are the slopes and intercepts of the two fitted lines identical (between the 2 years)? Would you use the 2022 model to predict for 2023?

3. Based on the R-squared values for the two cases, how do you think the models fit the data?

1. <br> For 2022: The regression was happiness_score = 4.5626 $\times$ healthy_life_expectancy + 2.8791
<br> For 2023: The regression was happiness_score = 5.4461 $\times$ healthy_life_expectancy + 3.5502
2. <br> The regressions are fairly different, for 2023 both the constant value and slope are significantly (more than the std err) larger than the 2022 regression values. This implies that using the 2022 model to predict values for 2023 would be a very poor choice.
3. <br> Because the R^2 (and adj. R^2) are quite low at 0.548 and 0.558 for each year respectively. This implies that there is a correlation between the prediction and data but it is fairly weak.

&#9989;  **Question 3.3 (6 points)**: Now make scatter plots of `happiness_score` (x-axis) vs. `healthy_life_expectancy` (on y-axis) for 2022 and 2023 on the same plot. Use different colors for the different years. Also plot the best fit lines on the same plot with same colors as the corresponding scatter plots. Label the axes, and add a legend, and give the plot a title.

```python
# Put your code here
fig, (ax22, ax23) = plt.subplots(2,1, figsize = (10,10)) 
#ax22.set_facecolor('k')
ax22.scatter(y_22,X_22['healthy_life_expectancy'], label = 'Measured Values', c = 'm')
ax22.plot(results_22.predict(), X_22['healthy_life_expectancy'], label = 'Predicted Values', c = 'k')
ax22.set_xlabel('happiness score')
ax22.set_ylabel('healty life expectancy')
ax22.set_title('2022 Data')
ax22.legend()

ax23.scatter(y_23,X_23['healthy_life_expectancy'], label = 'Measured Values', c = 'g')
ax23.plot(results_23.predict(), X_23['healthy_life_expectancy'], label = 'Predicted Values', c = 'c')
ax23.set_xlabel('happiness score')
ax23.set_ylabel('healty life expectancy')
ax23.set_title('2023 Data')
ax23.legend()
```

&#9989;  **Question 3.4 (2 points)**: In which year are the happiness score predictions generally better?

In 2023 the prediction is generally closer to the true values (indicated by a higher R^2 too), also because the 2023 data has more obvious few data points that are signficantly further away from the predicted value than most other data points unlike 2022 which is more globular.

**&#9989; Do this:** **Question 3.5 (4 points)**: Use `plot_regress_exog` to investigate the distribution of residuals in your model fits. Make sure to create a large enough figure so that everything is easily visible. Make separate plots for 2022 and 2023.

```python
# Put code here.
fig, ax = plt.subplots(figsize=(20, 15))
sm.graphics.plot_regress_exog(results_22, 1, fig=fig)
fig.suptitle('2022 Model', fontsize=32)
plt.show()
```

```python
# Put code here.
fig, ax = plt.subplots(figsize=(15, 15))
sm.graphics.plot_regress_exog(results_23, 1, fig=fig)

fig.suptitle('2023 Model', fontsize=32)
plt.show()
```

&#9989;  **Question 4.1 (5 points)**: Using the `OLS` method in `statsmodels`, perform a multivariable linear regression to predict the `happiness_score` based on `gdp_per_capita`, `healthy_life_expectancy`, `freedom_to_make_life_choices`, and `generosity`. Also, use the `add_constant()` method in `statsmodels` to ensure the model includes a constant term as well. Fit models to both the 2022 and 2023 data and display summaries of both results.

```python
# Put your code here
X_22_vector = sm.add_constant(whr_22_data[['healthy_life_expectancy','gdp_per_capita','freedom_to_make_life_choices','generosity']])
y_22 = whr_22_data['happiness_score']
X_23_vector = sm.add_constant(whr_23_data[['healthy_life_expectancy','gdp_per_capita','freedom_to_make_life_choices','generosity']])
y_23 = whr_23_data['happiness_score']


results_22_vector = sm.OLS(y_22, X_22_vector).fit()
results_23_vector = sm.OLS(y_23, X_23_vector).fit()

print(results_22_vector.summary())
print('-'*100)
print(results_23_vector.summary())
```
